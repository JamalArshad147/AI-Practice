```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
houses = pd.read_csv("kc_house_data.csv", nrows=10)
```

```python
houses.columns
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    houses["bedrooms"],
    houses["sqft_living"],
    color="y",
    linestyle="--",
    linewidth=2,
    marker="o",
    markersize=10,
    markerfacecolor="g",
    markeredgecolor="r",
    markeredgewidth=2,
)

plt.xlabel("Bedrooms")
plt.ylabel("Sqft_living")
plt.title("Bedrooms vs Sqft_living")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    houses["bedrooms"],
    houses["sqft_living"],
    color="y",
    marker="*",
)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    houses["bedrooms"],
    houses["sqft_living"],
    color="y",
    linestyle="--",
    linewidth=2,
    marker="o",
    markersize=10,
    markerfacecolor="g",
    markeredgecolor="r",
    markeredgewidth=2,
)

plt.bar(
    houses["bedrooms"],
    houses["sqft_living"],
    color="y",
    width=0.9,
)

plt.show()
```

```python
plt.figure(figsize=(5, 8))

plt.barh(
    houses["bedrooms"],
    houses["sqft_living"],
    color="y",
)

plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.hist(
    houses["bedrooms"],
    color="y",
    bins=20,
    alpha=0.5,
    orientation="horizontal"    
)

plt.xlim(5, 4)
plt.ylim(5, 4)

plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.boxplot(houses["bedrooms"], patch_artist=True)

plt.show()
```
